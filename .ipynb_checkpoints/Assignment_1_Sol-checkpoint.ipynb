# Assignment 1: Generating Molecules through Variational Autoencoders

## Autoencoders

Autoencoders are neuron-based generative models that utilize the concept of dimensionality reduction. Similar to methods like PCA or t-SNE, autoencoders use features to describe a complex set of input data. However, unlike PCA, where a clear recipe for the features must be provided, the autoencoder learns the best reduction method on its own. It achieves this by using a neural network to perform dimensionality reduction through a latent space. 

The process can be summarized as: $( x \rightarrow e(x) = z )$. Using this latent space, the model can then decode a vector $(z)$ back to its original dimension. This allows the autoencoder to directly compare the input $(x)$ with the reconstructed output $( d(e(x)))$, training itself in an unsupervised manner by minimizing the difference between them. A common way to measure this difference is by using the cross-entropy loss function.

#### The goal of such a model is to establish a bottleneck format of information propagation from input to output, retaining the most relevant information for reconstruction.


## Variational Autoencoders

As one might be able to derive from the previously shown architecture, the autoencoder model is prone to some issues. Due to the simplified approach of the latent space, this model can have a tendency towards overfitting data. 

If this concept sounds foreign, try to imagine the best working autoencoder possible. This autoencoder is able to transcribe all input $x = [x_1,x_2,...,x_n]$ towards a unique latent variable $(z)$. However, as you might realize, although such a model accomplishes perfect compression and decompression, it fails to accurately extract any useful information and thus fails to compress and decompress similar traits from unseen data. 

We therefore aim for a model that can analyze latent space representations and extract new information from this about the data. This is usually done through adding regularization in the autoencoder, creating a solution that is particularly useful as a generative model: the Variational Autoencoder (VAE).


#### Sampling from a distribution

The strength behind VAE's lies in the fact that instead of an exact point $z$ is sampled from the latent space, a norml distribution of points is sampled  $( \mathcal{N}(\mu, \sigma)$. This encoder output is comprimised of two vectors representing the mean $\mu$ and the standard deviation $\sigma$. The input for the decoder is then sampled from a distribution described by $( \mathbf{z} \sim \mathcal{N}(\mu, \sigma) )$. The training process is therefore comprimised of reconstructing this input and comparing it to the original one.


However, in comparison to the earlier defined standard loss function, a regularization term must be added. This is to move the distribution closer towards an actual standard normal distribution $( \mathcal{N}(0, 1) )$ to create a more continous and organized output. To measure this standard deviation of the normal distribution, the Kullback-Leibler (KL) divergence is often used. 

The introduction of noise through sampling serves as a direct regularization mechanism which works similar to dropout in neural networks. As often the case, regularization helps the model prevent overfitting but is also crucial for the production of new datapoints.

## Data Preprocessing
#### Garbage in = Garbage out

In this section we will prepare a dataset to be used in the variational autoencoder. As with most machine learning models, your data might not be completely clean and might contain bad data. We can therefore try and utilize RDkit's `MolFromSmiles()` Function to determine whether our dataset contains only valid smiles.


Besides this, we want our model to be able to have a target vectors to aim for. If the model has a target, the model can create a better latent space definition of the important bits within our data. Use the RDKit functions `Descriptiors.MolLogP()` and `QED.qed()` to generate two target columns defining the partition coefficient and the Quantative Estimate of Drug-likeliness. 

As an output, use the pandas library to create a dataframe with 3 columns. __SMILES__, __LogP__, __QED__. This will be required for our model to properly handle the data later on.

```python
def preprocess_smiles(input_file):
    # Read the original SMILES file
    smiles_data = pd.read_csv(input_file, usecols=[0])
    
    # Prepare lists to store valid SMILES, LogP, and QED
    valid_smiles = []
    logp_list = []
    qed_list = []
    total_smiles = len(smiles_data)  # Total number of SMILES in the input file

    for sm in tqdm(smiles_data.iloc[:, 0].dropna().tolist(), desc="Processing SMILES"):
        mol = Chem.MolFromSmiles(sm)
        if mol:  # Check if the molecule is valid
            valid_smiles.append(sm)
            logp = Descriptors.MolLogP(mol)
            qed = QED.qed(mol)
            logp_list.append(logp)
            qed_list.append(qed)
    
    # Create a DataFrame with the valid SMILES and their properties
    processed_data = pd.DataFrame({
        'SMILES': valid_smiles,
        'LogP': logp_list,
        'QED': qed_list
    })
    
    # Save the new data to a CSV file
    processed_data.to_csv('data/preprocessed_smiles.csv', index=False)
    print(f"Preprocessed data saved to {output_file}")
    
    # Calculate and print the percentage of valid SMILES
    valid_percentage = (len(valid_smiles) / total_smiles) * 100
    print(f"Percentage of valid SMILES: {valid_percentage:.2f}%")

# Test your function
preprocess_smiles('data/bad_data.smi')
```

For the upcoming section, the code will transform the dataset towards a representation that works best for the model. This is done through tokenizing the data such that each atom represents a token. You might have already spotted the earlier definition of the `_special_` dictionary. This represents the special tokens `{0: "PAD", 1: "BOS", 2: "EOS"}`, which represent the padding, beginning and end of a tokenized smile. The model works through processing batches of such representations and requires every batch to be equal in size. It therefore adds a Beginning of Sequence at the start, End Of Sequence at the end and PADding to fill them to equal sizes.

Since this section is mostly data handling, you are not required to make any adjustments. However, if you're familiar with object-based programming, feel free to scim through the class to get an idea of its functionality.

```python
class Smiles_data(torch.utils.data.Dataset):
    def __init__(self, file, total=130):
        self.total = total
        
        # Load the preprocessed dataset with SMILES and target variables
        data = pd.read_csv(file).iloc[:self.total]
        
        self.smiles = data['SMILES'].tolist()
        self.logp_qed = data[['LogP', 'QED']].values.tolist()
        
        tokens = functools.reduce(lambda acc, s: acc.union(set(s)), self.smiles, set())
        self.vocsize = len(tokens) + len(__special__)
        self.index2token = dict(enumerate(tokens, start=3))
        self.index2token.update(__special__)
        self.token2index = {v: k for k, v in self.index2token.items()}
        
        self.ints = [torch.LongTensor([self.token2index[s] for s in line]) for line in tqdm(self.smiles, "Preparing the dataset")]

    def decode(self, indexes):
        return "".join([self.index2token[index] for index in indexes if index not in __special__])

    def __getitem__(self, i):
        special_added = torch.cat((
            torch.LongTensor([self.token2index['<BOS>']]),
            self.ints[i],
            torch.LongTensor([self.token2index['<EOS>']]),
        ), dim=0)
        
        logp, qed = self.logp_qed[i]  # Get the LogP and QED values for the current index
        return one_hot(special_added, self.vocsize).float(), special_added, torch.tensor([logp, qed], dtype=torch.float32)

    def __len__(self):
        return len(self.smiles)


def pad_collate_fn(batch):
    seq_len = max([item[1].size(0) for item in batch])
    padded_batch = []
    properties = []

    for one_hot_seq, special_added, props in batch:
        pad_length = seq_len - special_added.size(0)
        padded_special_added = torch.cat([
            special_added,
            torch.LongTensor([0] * pad_length)
        ])
        padded_one_hot_seq = one_hot(padded_special_added, one_hot_seq.size(1)).float()
        padded_batch.append((padded_one_hot_seq, padded_special_added))
        properties.append(props)
    
    one_hot_tensors = torch.stack([item[0] for item in padded_batch])
    special_added_tensors = torch.stack([item[1] for item in padded_batch])
    properties = torch.stack(properties)
    
    return one_hot_tensors, special_added_tensors, properties
```

### The Variational Autoencoder (VAE) Model

## SmilesVAE Flowchart Structure

The `SmilesVAE` class follows the same flow as our earlier described scenario: **Input → Encoder → Latent Space → Decoder → Output**. The following section will describe how each component fits into this flow. Your job is to fill in the blanks in the code structure afterwards. Dont worry if you are not too familiar with the object based environment, if you keep the flowchart in mind, you can view each part as a cog in a system.

### Initialization (`__init__` method)
- **Purpose:** This is how a "class" sets up the layers and components of the VAE model. It defines which components are known throughout the class environment. They can be called through `"self.component."`
- **Components Initialized:**
  - **Encoder LSTM (`self.encoder_lstm`):** Processes the input sequence to generate "hidden states".
  - **Fully Connected Layers (`self.fc_mu` and `self.fc_logvar`):** Map the hidden state to the latent space by generating the mean (`mu`) and log-variance (`logvar`).
  - **Decoder LSTM (`self.decoder_lstm`):** Takes the latent vector and generates a sequence to reconstruct the input.
  - **Output Layer (`self.fc_out`):** Maps the hidden states of the decoder back to the vocabulary size based on our earlier defined  mean (`mu`) and log-variance (`logvar`).
  - **Predictor (`self.predictor`):** A small neural network to predict properties (like LogP and QED) from the latent vector.

### 1. Input
- **Data:** A sequence of tokens representing molecules in SMILES format.
- **Function:** The input sequence `x` is passed to the `forward` method.
- **Role in Flowchart:** Serves as the starting point for the VAE. The data is then passed to the encoder.

### 2. Encoder 
- **Function:** `encode(x)`
- **Process:**
  - Takes the input sequence and processes it using the encoder LSTM (`self.encoder_lstm`) to get a hidden representation.
  - This hidden state is passed through two fully connected layers (`self.fc_mu` and `self.fc_logvar`) to get the mean (`mu`) and log-variance (`logvar`).
- **Output:** `mu` and `logvar` vectors, which describe the latent distribution.
- **Role in Flowchart:** Encodes the input into a smaller, compressed form.

### 3. Latent Space 
- **Function:** `reparameterize(mu, logvar)`
- **Process:**
  - Uses the mean and log-variance to sample a latent vector `z` using the reparameterization trick. This involves creating a random vector that is adjusted by the mean and standard deviation derived from `logvar`.
- **Output:** The sampled latent vector `z`.
- **Role in Flowchart:** Represents the compact, encoded version of the input data.

### 4. Decoder
- **Function:** `decode(z, seq_len=10, tokenizer=None, temp=1.0)`
- **Process:**
  - Takes the latent vector `z` and tries to reconstruct the original input sequence.
  - Starts with a special token (`<BOS>`) and generates the sequence step-by-step using the decoder LSTM (`self.decoder_lstm`).
  - Uses `self.fc_out` to map LSTM outputs back to the vocabulary space to get logits for each token in the sequence.
- **Output:** The generated sequence and the logits.
- **Role in Flowchart:** Converts the latent representation back into a sequence.

### 5. Output
- **Function:** `forward(x, tokenizer, seq_len=10)`
- **Process:**
  1. **Encoding:** Calls `encode(x)` to compress the input sequence into `mu` and `logvar`.
  2. **Reparameterization:** Uses `reparameterize(mu, logvar)` to sample the latent vector `z`.
  3. **Decoding:** Calls `decode(z, seq_len, tokenizer)` to generate the reconstructed sequence from `z`.
  4. **Prediction:** Uses the `predictor` network to estimate molecular properties like LogP and QED from `z`.
- **Output:** Returns:
  - `logits`: The predicted sequence.
  - `mu` and `logvar`: Parameters of the latent distribution.
  - `predicted_properties`: Predicted properties of the input sequence.
- **Role in Flowchart:** Defines the entire data flow from input to output, combining encoding, latent space sampling, decoding, and property prediction.


```python
class SmilesVAE(nn.Module):
    def __init__(self, vocab_size, latent_dim=20, hidden_size=64):
        super(SmilesVAE, self).__init__()
        self.latent_dim = latent_dim
        self.hidden_size = hidden_size
        
        # Encoder
        self.encoder_lstm = nn.LSTM(vocab_size, hidden_size, batch_first=True)
        self.fc_mu = nn.Linear(hidden_size, latent_dim)
        self.fc_logvar = nn.Linear(hidden_size, latent_dim)
        
        # Decoder
        self.decoder_lstm = nn.LSTM(latent_dim, hidden_size, batch_first=True)
        self.fc_out = nn.Linear(hidden_size, vocab_size)
        
        # Predictor for LogP and QED (2 outputs)
        self.predictor = nn.Sequential(
            nn.Linear(latent_dim, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, 2)  # 2 outputs for LogP and QED
        )

    def encode(self, x):
        _, (hidden, _) = self.encoder_lstm(x)  # Learns hidden representation of the input vector x
        mu = self.fc_mu(hidden[-1])  # Use hidden representation to determine the center of the latent distribution (mean)
        logvar = self.fc_logvar(hidden[-1])  # Use hidden representation to determine the spread of the latent distribution (variance)
        return mu, logvar
    
    def reparameterize(self, mu, logvar):  # To sample from the distribution and ensure differentiability, use the reparameterization trick
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        z = mu + eps * (std + 1e-10)
        return z  # z is the sampled latent vector
    
    def decode(self, z, seq_len=10, tokenizer=None, temp=1.0):
        batch_size = z.size(0)
        generated_seq = torch.full((batch_size, 1), tokenizer['<BOS>'], dtype=torch.long, device=z.device)
        hidden_state = None

        logits_list = []

        for t in range(seq_len):
            z_rep = z.unsqueeze(1).repeat(1, generated_seq.size(1), 1)
            output, hidden_state = self.decoder_lstm(z_rep, hidden_state)
            logits = self.fc_out(output[:, -1, :])  # Get logits for the last time step
            logits_list.append(logits)

            next_token_probs = torch.softmax(logits / temp, dim=-1)
            next_token = torch.multinomial(next_token_probs, 1)

            generated_seq = torch.cat((generated_seq, next_token), dim=1)

        logits = torch.stack(logits_list, dim=1)

        # Retrieve <EOS> and <PAD> token indices from tokenizer
        eos_token = tokenizer['<EOS>']
        pad_token = tokenizer['<PAD>']

        # Stop generation after EOS token is encountered
        mask = torch.cumsum((generated_seq == eos_token).float(), dim=1)
        generated_seq = generated_seq.masked_fill(mask > 0, pad_token)

        return logits, generated_seq

    def forward(self, x, tokenizer, seq_len=10):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        logits, _ = self.decode(z, seq_len=seq_len, tokenizer=tokenizer)
        
        # Predict LogP and QED
        predicted_properties = self.predictor(z)
        
        return logits, mu, logvar, predicted_properties
```

### Loss Functions in the VAE

- `Reconstruction Loss`:
    - Measures how well the decoder reconstructs the original input sequence from the latent representation using Cross-Entropy.

    \begin{align*}
    \mathcal{L}_{\text{reconstruction}} = -\sum_{i=1}^{N} \log P(x_i | z)
    \end{align*}

- `KL Divergence Loss`:
    - Encourages the learned latent distribution to be close to a standard normal distribution, $( \mathcal{N}(0, 1))$.

    \begin{align*}
    \mathcal{L}_{\text{KL}} = -\frac{1}{2} \sum_{j=1}^{J} \left( 1 + \log \sigma_j^2 - \mu_j^2 - \sigma_j^2 \right)
    \end{align*}

- `Property Prediction Loss`:
    - Measures the accuracy of the predicted properties (LogP and QED) compared to the true properties using Mean Squared Error (MSE).

    \begin{align*}
    \mathcal{L}_{\text{property}} = \frac{1}{N} \sum_{i=1}^{N} \left( (\hat{y}_{\text{logP}, i} - y_{\text{logP}, i})^2 + (\hat{y}_{\text{QED}, i} - y_{\text{QED}, i})^2 \right)
    \end{align*}

### Total Loss in the VAE

The total loss function combines these three components: reconstruction loss, KL divergence, and property loss.


\begin{align*}
\mathcal{L}_{\text{total}} = \mathcal{L}_{\text{reconstruction}} + (\beta \cdot \mathcal{L}_{\text{KL}}) + \mathcal{L}_{\text{property}}
\end{align*}

Where:
- ( $\mathcal{L}_{\text{reconstruction}}$ ) is the cross-entropy loss.
- ( $\mathcal{L}_{\text{KL}}$ ) is the KL divergence loss.
- ( $\mathcal{L}_{\text{property}}$ ) is the MSE loss for predicting properties.
- ( $\beta $) is a weighting factor that controls the influence of the KL divergence, typically annealed over training to stabilize the learning process.


#### The following segment describes the loss function of the system. However, some blanks have been left in! Use your newfound knowledge about this system's loss function to fill them in and complete the code. 

(HINT: Look into and use the functions `nn.MSELoss()` and `nn.CrossEntropyLoss`.)

```python
def vae_loss_function(reconstructed_seq, original_seq, mu, logvar, predicted_properties, true_properties, epoch, n_epochs):
    # Flatten the tensors for CrossEntropyLoss
    flat_reconstructed_seq = reconstructed_seq.view(-1, reconstructed_seq.size(-1))  # (batch_size * seq_len, vocab_size)
    flat_original_seq = original_seq.view(-1)  # (batch_size * seq_len)
    
    # Define a mask for the padding sequences
    pad_mask = original_seq != 0  
    
    # Ensure reconstruction loss uses float logits and long target
    reconstruction_loss = nn.CrossEntropyLoss(reduction='none')(flat_reconstructed_seq, flat_original_seq.long())
    
    # Construct the loss without the padding mask
    reconstruction_loss = reconstruction_loss[pad_mask.view(-1)].mean()
    
    # KL Divergence
    kl_divergence = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())

    # Predictor loss (MSE Loss)
    property_loss = nn.MSELoss()(predicted_properties, true_properties)

    # Define beta for KL annealing
    beta = min(1.0, epoch / (n_epochs * 0.1))  # KL annealing

    # Construct the total loss
    total_loss = reconstruction_loss + (beta * kl_divergence) + property_loss
    
    return total_loss, reconstruction_loss.item(), kl_divergence.item(), property_loss.item()
```

### Training the Model

Lastly, all our earlier efforts come together in one final function that creates the model and trains it towards its target vectors. Once again, fill in the blanks in this section such that all the previous functions are used in their proper way. Try and look closely to what inputs and outputs these functions require. <br>
Dont worry about actually training the model, as long as it enters its first epoch your code is fine. 

```python
def train_vae(train_file, batch_size=32, learning_rate=0.0005, n_epochs=50, latent_dim=40, hidden_size=128, subset_size=50):
    save_file = f"vae_pred_subset{subset_size}_latent{latent_dim}_hidden{hidden_size}_epochs{n_epochs}.pt"
    
    # Call upon a function for your dataset
    dataset = Smiles_data(train_file, total=subset_size)

    # Call upon a function to generate your model
    model = SmilesVAE(dataset.vocsize, latent_dim, hidden_size).to(device)
    
    #Use torch.utils.data.DataLoader to wrap the dataset into batched with padding
    dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True, collate_fn=pad_collate_fn)

    # Enfore an optimizer to adjust the model's parameters during training
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

    epoch_reconstruction_losses = []
    epoch_kl_divergences = []
    epoch_property_losses = []

    for epoch in range(1, n_epochs + 1):
        total_loss = 0.
        total_reconstruction_loss = 0.
        total_kl_divergence = 0.
        total_property_loss = 0.

        for iteration, (batch, target, true_properties) in enumerate(tqdm(dataloader, 'Training')):
            batch, target, true_properties = batch.to(device), target.to(device), true_properties.to(device)
            model.train()
            seq_len = target.size(1)  # Get the sequence length from the target sequence
            reconstructed_seq, mu, logvar, predicted_properties = model(batch, dataset.token2index, seq_len=seq_len)

            # Call upon the earlier defined training function
            loss, reconstruction_loss, kl_divergence, property_loss = vae_loss_function(
                reconstructed_seq, target, mu, logvar, predicted_properties, true_properties, epoch, n_epochs
            )

            total_loss += loss.item()
            total_reconstruction_loss += reconstruction_loss
            total_kl_divergence += kl_divergence
            total_property_loss += property_loss

            # Use the loss.backwards to calculate the gradients and then updates the model
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        epoch_reconstruction_losses.append(total_reconstruction_loss / len(dataloader))
        epoch_kl_divergences.append(total_kl_divergence / len(dataloader))
        epoch_property_losses.append(total_property_loss / len(dataloader))

        print(f"Epoch {epoch} of {n_epochs} done, total loss: {total_loss}")

    torch.save({'tokenizer': dataset.index2token, 'model': model.cpu()}, f"models/{save_file}")
    print(f"VAE Training Complete! Model saved as {save_file}")

## Call upon the train_vae() function in order to train the model
subset_size = 100000
train_vae(train_file='processed_smiles_22.csv', batch_size=128, n_epochs=5, latent_dim=50, hidden_size=256, subset_size=subset_size)
```

## Generating Molecules

If you have reached this part of the notebooks I want to congratulate you! You should now have a beginning sense what it takes to create and train a LSTM-based Variational Autoencoder. However, we haven't even discussed one of the main talking points of such a model, namely it's generative power. <br>

Besides dimensionality reduction, the latent space representation of the original data allows data to be sampled from it as well! In this section, we will be exploring this method. 

```python
# Function to decode token indices into SMILES
def decode_from_indices(token_indices, tokenizer):
    # Join tokens into a string and remove any unwanted special tokens
    smiles = "".join([tokenizer.get(i, '') for i in token_indices if i not in __special__])
    
    # Remove any unwanted characters like tabs or newlines from the generated string
    smiles = smiles.replace('\t', '').replace('\n', '').strip()
    
    return smiles


def validate_smiles(smiles):
    ### Use the Chem.MolFromSmiles() function to validate your SMILES Strings
    ### The output should be True or False.
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            return True
    except:
        print(f"SMILES parsing failed: {smiles}")
        return False

def aim_for_target_properties(model, z, target_logp, target_qed, lr=0.01, steps=10):
    z = z.clone().detach().requires_grad_(True)
    optimizer = torch.optim.Adam([z], lr=lr)
    
    for _ in range(steps):
        optimizer.zero_grad()
        predicted_properties = model.predictor(z)
        predicted_logp, predicted_qed = predicted_properties[0]

        # Loss to guide z towards the target properties
        loss = ((predicted_logp - target_logp) ** 2 + (predicted_qed - target_qed) ** 2)
        loss.backward()
        optimizer.step()
    
    return z.detach()

def generate_molecules(model, tokenizer, num_samples, seq_len, target_logp=None, target_qed=None, temp=1.0, max_temp=1.05, max_attempts=15):
    model.eval()  # Set model to evaluation mode
    
    generated_smiles = []
    unique_smiles = set()  # Set to track unique SMILES
    token2index = {v: k for k, v in tokenizer.items()}  # Reverse tokenizer mapping
    
    try:
        eos_token_index = token2index['<EOS>']  # Retrieve the index for the <EOS> token
        print(f"<EOS> token found in tokenizer with index: {eos_token_index}")
    except KeyError:
        raise KeyError("The '<EOS>' token is missing in the tokenizer.")
    
    initial_temp = temp  
    sample_idx = 0
    attempts = 0  
    ## Use a simple loop to start the search. The search must adjust the temperature based on attemps. 
    ### Use the variables "temp","max_attemps" and "max_temp" in your loop.
    while sample_idx < num_samples:
        if attempts >= max_attempts:  
            new_temp = min(temp * 1.05, max_temp)  # Increase temperature but cap it
            if new_temp != temp:
                temp = new_temp
                print(f"Increasing temperature to {temp} for more diversity")
            attempts = 0 
        
        # Use the temperature and torch.randn() to sample from the distribution for the variable z.
        # (Dont forget to implement the temperature!)
        z = torch.randn(1, model.latent_dim) * temp  # Shape: (1, latent_dim)

        # Check if we have target values to aim for
        if target_logp is not None and target_qed is not None:
            # Iteratively adjust z to aim for the target properties (e.g., through gradient descent)
            z = aim_for_target_properties(model, z, target_logp, target_qed)

        
        # Decode the latent vector to generate logits and the token sequence
        logits, sampled_tokens = model.decode(z, seq_len=seq_len, tokenizer=token2index, temp=temp)  # Pass token2index
        
        # Convert token indices back to SMILES string
        token_indices = sampled_tokens[0].tolist()  # Take the first (and only) sample from batch
        if eos_token_index in token_indices:
            token_indices = token_indices[:token_indices.index(eos_token_index)]  # Stop at <EOS>
        
        smiles = decode_from_indices(token_indices, tokenizer)
        
        # Validate SMILES string
        if validate_smiles(smiles):
            if smiles not in unique_smiles:
                print(f"Valid unique SMILES: {smiles}")
                generated_smiles.append(smiles)
                unique_smiles.add(smiles)
                sample_idx += 1
                attempts = 0  # Reset attempts counter after successful generation
                temp = initial_temp  # Reset temperature to initial value
            else:
                attempts += 1  # Increment attempts
        else:
            attempts += 1  # Increment attempts for invalid SMILES
    
    return generated_smiles
```

### Calling the function

```python
# Example usage
if __name__ == "__main__":
    # Load the trained model and tokenizer
    checkpoint = torch.load('models/LSTM_VAE.pt')
    model = checkpoint['model']
    tokenizer = checkpoint['tokenizer']
    
    # Define target properties (optional)
    target_logp = 2.0
    target_qed = 0.6
    
    # Generate molecules
    num_samples = 10
    temperature = 0.55
    seq_len = 11
    generated_smiles = generate_molecules(model, tokenizer, num_samples, seq_len, target_logp=target_logp, target_qed=target_qed, temp=temperature)

    # Save generated SMILES to a file
    df = pd.DataFrame(generated_smiles, columns=["SMILES"])
    df.to_csv(f"generated_molecules_len{seq_len}.csv", index=False)
    
    print(f"Generated {len(generated_smiles)} valid molecules and saved to generated_molecules.csv")
    print(df)
```

## Bonus: Visualizing Molecules

The library rdkit has some useful functions that can be used for the visualization of your generated molecules! Besides this being fun, it can also help better determine whether your model has any bias within itself. E.g. Having a model that prefers generating long series of C.

Once again, use the rdkit function `Draw.MolsToGridImage()` to generate a drawing of one or multiple of your model's molecules.

```python
def visualize_smiles_molecules(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Extract SMILES from the first column
    smiles_list = df.iloc[:, 0].tolist()
    
    # Convert SMILES to RDKit molecule objects
    mols = [Chem.MolFromSmiles(smiles) for smiles in smiles_list if smiles]
    
    # Visualize the molecules in a grid
    img = Draw.MolsToGridImage(mols, molsPerRow=4, subImgSize=(200, 200), legends=[f"Mol {i}" for i in range(len(mols))])
    
    # Display the image directly in the notebook
    display(img)

img = visualize_smiles_molecules('molecules/generated_molecules_len17.csv')
```
